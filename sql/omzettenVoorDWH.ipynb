# DimDate

```python
import pandas as pd
from datetime import datetime, timedelta

# Bereken de startdatum en de einddatum
start_date = datetime(1960, 1, 1)
end_date = datetime(datetime.now().year, datetime.now().month, 1) + pd.DateOffset(months=1) - timedelta(days=1)

# Maak een range van datums van start_date tot end_date
date_range = pd.date_range(start=start_date, end=end_date)

# Creëer een DataFrame met de DateKey kolom
dimDate = pd.DataFrame({
    'DateKey': date_range.strftime('%Y%m%d')
})

# Bekijk de eerste paar rijen van de DataFrame
dimDate.head()
```

# DimTime

```python
time_keys = []
time_displays = []
Uur = []
Minuut = []

time_key = 1
while time_key <= 24 * 60:
    hour = (time_key - 1) // 60
    minute = (time_key - 1) % 60
    time_display = f"{hour:02d}:{minute:02d}"
    
    key = hour*100+minute
    time_keys.append(key)
    time_displays.append(time_display)
    Uur.append(hour)
    Minuut.append(minute)
    
    time_key += 1

dimTime = pd.DataFrame({'TimeKey': time_keys, 'Tijdstip': time_displays, 'Uur': Uur, 'Minuut': Minuut})

dimTime.head()
```

# DimTeam

```python
columns = ['jaar', 'maand', 'dag', 'uur', 'minuten', 'thuisploeg', 'uitploeg', 'scoreThuis', 'scoreUit', 'type']
dimTeam = pd.read_csv('../data/csv/opgekuist/wedstrijdenFlashScore.csv', names=columns, encoding='utf-8')
dimTeam.head()
```

```python
dimTeam = dimTeam[['thuisploeg', 'uitploeg']]
thuis = dimTeam['thuisploeg'].value_counts()
thuis = thuis.reset_index()
thuis.columns = ['ploeg', 'uit_count']
thuis.head()
```

```python
uit = dimTeam['uitploeg'].value_counts()
uit = uit.reset_index()
uit.columns = ['ploeg', 'uit_count']
uit.head()
```

```python
dimTeam = pd.merge(thuis, uit, on='ploeg', how='outer')
dimTeam = dimTeam[['ploeg']]
dimTeam['key'] = dimTeam.index
dimTeam.head()
```

# DimType

```python
columns = ['jaar', 'maand', 'dag', 'uur', 'minuten', 'thuisploeg', 'uitploeg', 'scoreThuis', 'scoreUit', 'type']
dimType = pd.read_csv('../data/csv/opgekuist/wedstrijdenFlashScore.csv', names=columns, encoding='utf-8')
dimType.head()
```

```python
dimType = dimType[['type']]
dimType = dimType['type'].value_counts()
dimType = dimType.reset_index()
dimType.columns = ['type', 'uit_count']
dimType = dimType[['type']]
dimType['key'] = dimType.index
dimType.head()
```

# FactWedstrijd

```python
columns = ['jaar', 'maand', 'dag', 'uur', 'minuten', 'thuisploeg', 'uitploeg', 'scoreThuis', 'scoreUit', 'type']
factWedstrijd = pd.read_csv('../data/csv/opgekuist/wedstrijdenFlashScore.csv', names=columns, encoding='utf-8')
factWedstrijd.head()
```

```python
factWedstrijd['jaar'] = factWedstrijd['jaar'].astype(str)
factWedstrijd['maand'] = factWedstrijd['maand'].astype(str).str.zfill(2)
factWedstrijd['dag'] = factWedstrijd['dag'].astype(str).str.zfill(2)
factWedstrijd['uur'] = factWedstrijd['uur'].astype(str).str.zfill(2)
factWedstrijd['minuten'] = factWedstrijd['minuten'].astype(str).str.zfill(2)

# Combineer de kolommen om dateKey en timeKey te vormen
factWedstrijd['dateKey'] = factWedstrijd['jaar'] + factWedstrijd['maand'] + factWedstrijd['dag']
factWedstrijd['timeKey'] = factWedstrijd['uur'] + factWedstrijd['minuten']
factWedstrijd.head()
```

```python
factWedstrijd = factWedstrijd[['dateKey', 'timeKey', 'thuisploeg', 'uitploeg', 'scoreThuis', 'scoreUit', 'type']]
factWedstrijd.head()
```

```python
factWedstrijd = factWedstrijd.merge(dimTeam, how='left', left_on='thuisploeg', right_on='ploeg')
factWedstrijd['homeTeamKey'] = factWedstrijd['key']
factWedstrijd = factWedstrijd.drop(['key', 'thuisploeg', 'ploeg'], axis=1)
factWedstrijd.head()
```

```python
factWedstrijd = factWedstrijd.merge(dimTeam, how='left', left_on='uitploeg', right_on='ploeg')
factWedstrijd['awayTeamKey'] = factWedstrijd['key']
factWedstrijd = factWedstrijd.drop(['key', 'uitploeg', 'ploeg'], axis=1)
factWedstrijd.head()
```

```python
factWedstrijd = factWedstrijd.merge(dimType, how='left', on='type')
factWedstrijd['typeKey'] = factWedstrijd['key']
factWedstrijd = factWedstrijd.drop(['key', 'type'], axis=1)
factWedstrijd.head()
```
